```python
from pythainlp.tokenize import word_tokenize
from transformers import AutoTokenizer
import deepcut 


tokenizer = AutoTokenizer.from_pretrained("openthaigpt/openthaigpt-1.0.0-7b-chat")

tokenizer.add_tokens(["ลาบูบู้", "เสริมศักดิ์", "พงษ์พานิช", "LABUBU", "POP", "MART", "Ceremony"])
```

```python
from time import time
att_start_time = time()
att_token = word_tokenize(txt, engine="attacut", custom_dict=["ลาบูบู้", "เสริมศักดิ์", "พงษ์พานิช", "LABUBU", "POP", "Ceremony"])
att_end_time = time()

newmm_start_time = time()
newmm_token = word_tokenize(txt, engine="newmm")
newmm_end_time = time()

newmm_safe_start_time = time()
newmm_safe_token = word_tokenize(txt, engine="newmm-safe")
newmm_safe_end_time = time()

longest_start_time = time()
longest_token = word_tokenize(txt, engine="longest")
longest_end_time = time()

pythai_start_time = time()
pythai_token = word_tokenize(txt)
pythai_end_time = time()

deep_start_time = time()
deep_token = deepcut.tokenize(txt, custom_dict=["ลาบูบู้", "เสริมศักดิ์", "พงษ์พานิช"])
deep_end_time = time()

huggingface_start_time = time()
huggingface_token = tokenizer.batch_decode(tokenizer(txt)["input_ids"][1:])
huggingface_end_time = time()


att_time = att_end_time - att_start_time
newmm_time = newmm_end_time - newmm_start_time
newmm_self_time = newmm_safe_end_time - newmm_safe_start_time
longest_time = longest_end_time - longest_start_time
pythai_time = pythai_end_time - pythai_start_time
deep_time = deep_end_time - deep_start_time
huggingface_time = huggingface_end_time - huggingface_start_time

# Print execution times
print("Execution Times:")
print(f"  - Attacut:                  {att_time:.4f} seconds", att_token)
print(f"  - NewMM:                    {newmm_time:.4f} seconds", newmm_token)
print(f"  - NewMM Safe:               {newmm_self_time:.4f} seconds", newmm_safe_token)
print(f"  - Longest Match:            {longest_time:.4f} seconds", longest_token)
print(f"  - PyThaiNLP:                {pythai_time:.4f} seconds", pythai_token)
print(f"  - DeepCut:                  {deep_time:.4f} seconds", deep_token)
print(f"  - Hugging Face Transformer: {huggingface_time:.4f} seconds", huggingface_token)
```
